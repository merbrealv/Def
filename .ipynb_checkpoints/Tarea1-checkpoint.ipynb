### 1.1. Objetivo del estudio

El objetivo del estudio consiste en estudiar el factor de transcripción LHY, factor de elongación tardío del hipocotilo de *Arabidopsis thaliana*. Este factor de transcripción forma parte del reloj circardiano. Se pretende estudiar las posibles interacciones del reloj circadiano con la biosíntesis y regulación del ácido abcísico (ABA) a través del factor LHY.

### 1.2. Diseño Experimental

Para el estudio se usó como sujeto de estudio *Arabidopsis thaliana*. Las muestras fueron extraídas dos horas después del amanecer cuando la expresión de LHY es máxima. A estas muestras se les realizó experimentos de inmunoprecipitación de la cromatina. Se llevarón a cabo con anticuerpos frente a todo el factor LHY. El ADN input fue usado como control en el experimento de ChIP. 

El número de acceso de GEO del estudio es GSE103785. Los SRA de cada muestra, chip e input, se encuentran en el fichero de parámetros del estudio. 
  	

### 1.3. Flujo de trabajo

El flujo de trabajo consiste en una sucesión de scripts automatizados. En primer lugar se realiza un análisis de chIP-Seq, donde se analiza la interacción de proteínas con el ADN. Esta técnica combina la inmunoprecipitación de la cromatina con la secuenciación masiva del ADN. Con el script pipe.sh podemos obtener los sitios de unión al ADN. 

A continuación, son lanzados los scripts de procesamiento de chip e input. Estos scripts realizan el análisis de calidad de las muestras, el mapeo y ensamblado de transcritos. Posteriormente, con otro script se obtienen los picos comparativos de las muestras para poder realizar los otros análisis. 

Con un script de R específico para el proyecto, se realiza un estudio funcional para conocer las funciones moleculares asociadas a estos picos. 

Finalmente, se usa la herramienta HOMER para determinar la familia a la que pertenece el factor de transcripción. 



# 2. Análisis de la calidad 

El estudio de la calidad de las muestras nos permitirá verificar nuestros resultados. De esta forma, podemos definir si nuestros resultados se deben verdaderamente a nuestras muestras o han existido contaminaciones.También, pueden existir problemas con las propias muestras presentando una calidad baja impidiendo la obtención de resultados concluyentes.  

Dicho estudio se realiza con **FastQC** proporcionándonos distintos tipos de gráficos de calidad. Todos estos gŕaficos se recogen en un único fichero y proporcionan la siguiente información: 

* **Estadísticos básicos** : Este parámetro permite conocer el fichero que ha sido procesado, el número de secuencias analizadas, la longitud de bases de cada secuencia y el porcentaje del contenido en GC. 



* **Calidad de bases por secuencia** : Este gráfico representa la calidad frente al número de bases totales. Es un gŕafico de tipo boxplot donde cada caja representa las lecturas que se han realizado. En función de la franja en la que se encuentren la calidad varía. Se debe destacar que la calidad decrece hacia el final debido a fallos producidos por la polimerasa. El color verde indica la zona de buena calidad y el rojo la zona de mala calidad.



* **Contenido de bases por secuencia** : Este gráfico presenta el contenido en bases. Se espera su correspondencia con el genoma del organismo. En el caso de *Arabidopsis thaliana* el contenido en AT es superior al contenido en GC. Si se obtienen otros porcentajes, este tipo de gráfico permite detectar la presencia de contaminaciones. La distorsión inicial se debe al adaptador de la secuenciación y su eliminación.

# 4. Enriquecimiento de motivos

Los motivos de unión de factores de transcripción son secuencias genómicas que especificamente unen a los factores de transcripción. Las secuencias consenso de estos motivos son variables, algunas bases son fijas otras no. Normalmente se presentan con un gráfico de tipo logo donde la altura del nucleótido indica la frecuencia. 
Los motivos suelen tener asociada una puntuación que indica la probabilidad de aparecer del motivo y la capacidad de unión del factor de transcripción. 

El enriquecimiento de motivos para nuestro estudio se realiza con la herramienta **Homer**. A continuación, presentamos los tres primeros motivos obtenidos: 


En la imagen anterior, encontramos el primer lugar el diagrama de bases de cada motivo. Podemos comprobar cuales son las bases más frecuentes en cada motivo. Otros datos interesantes a destacar son los p-valores de cada motivo y el porcentaje de dianas. Para el segundo y el tercer motivo encontramos el mismo valor de porcentaje de dianas y de p-valor. El motivo uno difiere con el resto sobre todo al estudiar el porcentaje de dianas.

Cuando estudiamos el primer motivo se relaciona con un motivo conocido, ABI3VP1, relacionado con el factor de transcripción RAV1. Este factor regula la senescencia de la planta y se incluye en la familia de factores de transcripción AP2/ERF implicada en el control del estrés abiotico en las plantas. Otro motivo, NTM2(NAC),se encuentra relacionado con el gen NAC69. Este gen codifica para un activador transcripcional implicado en la respuesta a estrés durante la germinación de las plantas y el crecimiento.


En el resto de motivos encontramos relaciones con genes implicados en las mismas funciones. Todo para finalmente, poder llegar a una conclusión final. La mayoría de resultados que encontramos se relacionan con genes implicados en la transcipción de factores implicados en activación e inhibición de otros genes. Por ejemplo, el motivo Q8LSZ4, se relaciona con el gen E2FE. Este gen codifica un factor de transcripción que inhibe la activación de genes.
